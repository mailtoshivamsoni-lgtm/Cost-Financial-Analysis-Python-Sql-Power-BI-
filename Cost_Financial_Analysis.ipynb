```python
import pandas as pd

df = pd.read_csv("Cost_Financial_Analysis.csv")
df.head()
```

```python
df.isnull().sum().sum()
```

```python
df.columns=df.columns.str.strip().str.replace(" ","_")
```

```python
df.Status=df.Status.str.replace("On going","Ongoing")
```

```python
df['Start_Date'] = pd.to_datetime(df['Start_Date'], errors='coerce', format='mixed', dayfirst=True)
```

```python
df['End_Date'] = pd.to_datetime(df['End_Date'], errors='coerce', format='mixed', dayfirst=True)
```

```python
df['Start_Date'] = df['Start_Date'].dt.strftime('%Y-%m-%d')
```

```python
df.Payment_Recvd = df.Payment_Recvd.fillna(0)
```

```python
df.Outstanding = df.Outstanding.fillna(0)
```

```python
df.Outstanding = df.Invoice_Amt - df.Payment_Recvd
```

```python
df['Cost_Variance'] = df['Actual_Cost'] - df['Budget_(INR)']
```

```python
df['Over_Budget_Flag'] = df['Cost_Variance'].apply(lambda x: "Yes" if x > 0 else "No")
```

```python
df = df.rename(columns={"Budget_(INR)": "Total_Budget","Actual_Cost": "Cost_Actual"})
```

```python
df.head()
```

```python
df.to_csv('New_Cost_Financial_Analysis.csv',index=False)
```
